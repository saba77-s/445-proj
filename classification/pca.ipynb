```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.rcParams['figure.dpi']= 200
from sklearn.decomposition import PCA

data = pd.read_csv("Breast_GSE70947.csv")
data = data.to_numpy()
y = data[:,1]
y = 1 * (y == 'normal')
X = data[:,2:]
```

```python
pca = PCA(n_components=2)
pca.fit(X)
data_pcaed = pca.transform(X)

class1_mask = (y == 0) # a boolean mask for class 1
class2_mask = (y == 1) # a boolean mask for class 2

sns.set(rc={"figure.figsize":(17,12)})
plt.scatter(data_pcaed[class1_mask,0], data_pcaed[class1_mask,1], c='b', label='Breast Cancer')
plt.scatter(data_pcaed[class2_mask,0], data_pcaed[class2_mask,1], c='r', label='No Breast Cancer')
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Top 2 Principal Components")
```
